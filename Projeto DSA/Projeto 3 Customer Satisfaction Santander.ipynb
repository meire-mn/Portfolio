DATA CLEANING

```python
###remove columns with uniform values (standard deviation=0)
uni=[]
for col in cols:
    if treino[col].std()==0:
        uni.append(col)
```

```python
treino.drop(uni, axis=1, inplace=True)
teste.drop(uni, axis=1, inplace=True)
```

```python
#remove duplicated columns
same=[]
colunas=treino.columns
for i in range(len(colunas)-1): 
    for j in range(i+1, len(colunas)):
        if treino[colunas[i]].equals(treino[colunas[j]]):
            same.append([colunas[j]])
```

```python
same=np.array(same)
unique_same=list(np.unique(same))
type(unique_same)
```

```python
treino.drop(unique_same, axis=1, inplace=True)
teste.drop(unique_same, axis=1, inplace=True)
```

```python
treino.shape
```

```python
# drop values -99999 from column 'var3' 
```

```python
treino.drop(treino.loc[treino['var3']==-999999].index, inplace=True)
```

```python
teste.drop(teste.loc[teste['var3']==-999999].index, inplace=True)
```

```python
treino.head()
```

```python
treino.shape
```

```python
#drop 'ID' column
treino.drop(['ID'], axis=1, inplace=True)
teste.drop(['ID'], axis=1, inplace=True)
```

```python
###normalize data
from sklearn.preprocessing import Normalizer

# Gerando os dados normalizados
X = treino.iloc[:,0:305]
Y = treino.iloc[:,306]
 
scaler = Normalizer().fit(X)
X = scaler.transform(X)
```

```python
#split 'treino' in train/test 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 10)
X_train.shape
```

```python
# Feature Selection with Extra Trees Classifier

modelo = ExtraTreesClassifier()
modelo.fit(X, Y)

# Select features
feat=SelectFromModel(modelo, prefit=True) #threshold (default=mean)

X_train=feat.transform(X_train)
```

```python
X_train.shape #70 variables
```

```python
X_test=feat.transform(X_test)
```

MODEL TRAINING

```python
#model training #Logistic Regression

mod_log=LogisticRegression()
mod_log.fit(X_train,Y_train)
```

```python
#predictions
predictions=mod_log.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, predictions))
```

```python
#ROC
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test,predictions)
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Acurácia: %.2f%%" % (accuracy * 100.0)) #too high #imbalanced data
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
mod_knn=KNeighborsClassifier()
mod_knn.fit(X_train,Y_train)
```

```python
#KNN predictions
pred_knn=mod_knn.predict(X_test)
```
